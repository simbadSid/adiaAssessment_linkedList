```python
from matplotlib import pyplot as plt
import math

plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True


SIZE                                    = [1000000, 11000000, 21000000, 31000000, 41000000, 51000000, 61000000, 71000000, 81000000, 91000000]

executionTime_0_orderRefactor_onlyForward            = [18, 193, 362, 529, 765, 1199, 1216, 1468, 2872, 2801]
executionTime_0_orderRefactor_onlyBackward           = [20, 208, 374, 557, 850, 1055, 1287, 2265, 3378, 6278]
executionTime_0_orderRefactor_halfForwardHalfBackward= [42, 288, 540, 868, 1184, 1502, 1826, 2831, 6524, 9394]
executionTime_0_orderRefactor_randomForwardBackward  = [26, 275, 511, 810, 1120, 3341, 3731, 6377, 5684, 11342]

executionTime_1_twoPointersInOn_onlyForward            = [21, 246, 455, 708, 925, 1210, 1477, 2607, 2060, 6703]
executionTime_1_twoPointersInOn_onlyBackward           = [20, 239, 457, 697, 923, 1160, 1403, 2952, 5427, 9176]
executionTime_1_twoPointersInOn_halfForwardHalfBackward= [24, 286, 944, 1283, 2776, 1387, 2613, 5337, 9827, 5725]
executionTime_1_twoPointersInOn_randomForwardBackward  = [25, 234, 450, 1387, 1279, 1237, 1790, 4573, 3497, 6744]

#################################
fig = plt.figure()
plt.grid()

plt.plot(SIZE, executionTime_0_orderRefactor_onlyForward,            label="0_orderRefactor_onlyForward",            color="green",  marker="o")
plt.plot(SIZE, executionTime_0_orderRefactor_onlyBackward,           label="0_orderRefactor_onlyBackward",           color="blue",   marker="o")
plt.plot(SIZE, executionTime_0_orderRefactor_halfForwardHalfBackward,label="0_orderRefactor_halfForwardHalfBackward",color="black",  marker="o")
plt.plot(SIZE, executionTime_0_orderRefactor_randomForwardBackward,  label="0_orderRefactor_randomForwardBackward",  color="purple", marker="o")

plt.plot(SIZE, [0.00001*i for i in SIZE],    '--', label="a*N (a=10^-5)",    color="red", marker="o")
plt.plot(SIZE, [0.00001*i*10 for i in SIZE], '--', label="a*N*10 (a=10^-5)", color="red", marker="o")

plt.xlabel('List size')
plt.ylabel('Total execution time (ms)')
plt.title('Execution time benchmark: implementation 0_orderRefactor')
plt.legend(loc='upper left')
plt.savefig('benchmark_0_orderRefactor.png')


#################################
fig = plt.figure()
plt.grid()

plt.plot(SIZE, executionTime_1_twoPointersInOn_onlyForward,            label="1_twoPointersInOn_onlyForward",            color="green",  marker="o")
plt.plot(SIZE, executionTime_1_twoPointersInOn_onlyBackward,           label="1_twoPointersInOn_onlyBackward",           color="blue",   marker="o")
plt.plot(SIZE, executionTime_1_twoPointersInOn_halfForwardHalfBackward,label="1_twoPointersInOn_halfForwardHalfBackward",color="black",  marker="o")
plt.plot(SIZE, executionTime_1_twoPointersInOn_randomForwardBackward,  label="1_twoPointersInOn_randomForwardBackward",  color="purple", marker="o")

plt.plot(SIZE, [0.00001*i for i in SIZE],    '--', label="a*N (a=10^-5)",    color="red", marker="o")
plt.plot(SIZE, [0.00001*i*10 for i in SIZE], '--', label="a*N*10 (a=10^-5)", color="red", marker="o")

plt.xlabel('List size')
plt.ylabel('Total execution time (ms)')
plt.title('Execution time benchmark: implementation 1_twoPointersInOn')
plt.legend(loc='upper left')
plt.savefig('benchmark_1_twoPointersInOn.png')


#################################
fig = plt.figure()
plt.grid()

plt.plot(SIZE, executionTime_0_orderRefactor_onlyForward,   label="0_orderRefactor_onlyForward",  color="green",  marker="o")
plt.plot(SIZE, executionTime_1_twoPointersInOn_onlyForward, label="1_twoPointersInOn_onlyForward",color="red",  marker="o")

plt.xlabel('List size')
plt.ylabel('Total execution time (ms)')
plt.title('Execution time benchmark: scenario onlyForward')
plt.legend(loc='upper left')
plt.savefig('benchmark_onlyForward.png')


#################################
fig = plt.figure()
plt.grid()

plt.plot(SIZE, executionTime_0_orderRefactor_onlyBackward,   label="0_orderRefactor_onlyBackward",   color="green",  marker="o")
plt.plot(SIZE, executionTime_1_twoPointersInOn_onlyBackward, label="1_twoPointersInOn_onlyBackward", color="red",  marker="o")

plt.xlabel('List size')
plt.ylabel('Total execution time (ms)')
plt.title('Execution time benchmark: scenario onlyBackward')
plt.legend(loc='upper left')
plt.savefig('benchmark_onlyBackward.png')


#################################
fig = plt.figure()
plt.grid()

plt.plot(SIZE, executionTime_0_orderRefactor_halfForwardHalfBackward,  label="0_orderRefactor_halfForwardHalfBackward",  color="green",  marker="o")
plt.plot(SIZE, executionTime_1_twoPointersInOn_halfForwardHalfBackward,label="1_twoPointersInOn_halfForwardHalfBackward",color="red",  marker="o")

plt.xlabel('List size')
plt.ylabel('Total execution time (ms)')
plt.title('Execution time benchmark: scenario halfForwardHalfBackward')
plt.legend(loc='upper left')
plt.savefig('benchmark_halfForwardHalfBackward.png')


#################################
fig = plt.figure()
plt.grid()

plt.plot(SIZE, executionTime_0_orderRefactor_randomForwardBackward,  label="0_orderRefactor_randomForwardBackward",  color="green",  marker="o")
plt.plot(SIZE, executionTime_1_twoPointersInOn_randomForwardBackward,label="1_twoPointersInOn_randomForwardBackward",color="red",  marker="o")

plt.xlabel('List size')
plt.ylabel('Total execution time (ms)')
plt.title('Execution time benchmark: scenario randomForwardBackward')
plt.legend(loc='upper left')
plt.savefig('benchmark_randomForwardBackward.png')







plt.show()

```
