```python
from matplotlib import pyplot as plt
import math

plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
fig = plt.figure()


SIZE                                    = [1000000, 11000000, 21000000, 31000000, 41000000, 51000000, 61000000, 71000000, 81000000, 91000000]

executionTime_0_orderRefactor_onlyForward            = [17, 214, 545, 801, 1045, 1235, 1574, 1898, 4173, 3214]
executionTime_0_orderRefactor_onlyBackward           = [32, 341, 658, 957, 1629, 1993, 2453, 2986, 5102, 5342]
executionTime_0_orderRefactor_halfForwardHalfBackward= [24, 273, 517, 765, 1257, 1834, 2198, 3259, 2926, 5510]
executionTime_0_orderRefactor_randomForwardBackward  = [53, 552, 1223, 1739, 2432, 2891, 4022, 6577, 7673, 12435]

plt.grid()

plt.plot(SIZE, executionTime_0_orderRefactor_onlyForward,            label="0_orderRefactor_onlyForward",            color="green",  marker="o")
plt.plot(SIZE, executionTime_0_orderRefactor_onlyBackward,           label="0_orderRefactor_onlyBackward",           color="blue",   marker="o")
plt.plot(SIZE, executionTime_0_orderRefactor_halfForwardHalfBackward,label="0_orderRefactor_halfForwardHalfBackward",color="black",  marker="o")
plt.plot(SIZE, executionTime_0_orderRefactor_randomForwardBackward,  label="0_orderRefactor_randomForwardBackward",  color="purple", marker="o")

plt.plot(SIZE, [0.00001*i for i in SIZE],    '--', label="a*N (a=10^-5)",    color="red", marker="o")
plt.plot(SIZE, [0.00001*i*10 for i in SIZE], '--', label="a*N*10 (a=10^-5)", color="red", marker="o")

plt.xlabel('List size')
plt.ylabel('Total execution time (ms)')
plt.legend(loc='upper left')

"""
plt.xscale('log')
plt.yscale('log')
plt.savefig('benchmark_log.png')
plt.title('Execution time benchmark (log scale)')
"""

plt.savefig('benchmark.png')
plt.title('Execution time benchmark')

plt.show()

```
